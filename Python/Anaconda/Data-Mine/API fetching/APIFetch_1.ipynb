<h2 style="text-align:center;font-size:40px">Libraries</h2>

```python
import pandas as pd
import requests
from tqdm.auto import tqdm
import time
```

<h2 style='text-align:center; font-size:40px'>Lists</h2>

```python
rank=[]
title=[]
title_english=[]
type_1=[]
type_2=[]
episodes=[]
source=[]
status=[]
airing=[]
duration=[]
rating=[]
score=[]
votes=[]
favourites=[]
season=[]
year=[]
genres=[]
themes=[]
demographics=[]
studio=[]
```

<h2 style="font-size:40px; text-align:center"> Fetching with API </h2>

```python
for p in tqdm(range(1,1202),desc='Fetching data...'):
    url='https://api.jikan.moe/v4/top/anime?&page={}'.format(p)
    response=requests.get(url).json()['data']

    for i in response:
    
        rank.append(i.get('rank') or 'NA')
        title.append(i.get('title') or 'NA')
        title_english.append(i.get('title_english') or 'NA')
        type_1.append('Series' if i.get('type')=='TV' else i.get('type') or 'NA')
        type_2.append(i.get('studios')[0].get('type') if i.get('studios') else 'NA')
        episodes.append(i.get('episodes') or 'NA')
        source.append(i.get('source') or 'NA')
        status.append(i.get('status') or 'NA')
        airing.append(i.get('airing') or 'NA')
        duration.append(i.get('duration') or 'NA')
        rating.append(i.get('rating') or 'NA')
        score.append(i.get('score') or 'NA')
        votes.append(i.get('scored_by') or 'NA')
        favourites.append(i.get('favorites') or 'NA')
        season.append(i.get('season') or 'NA')
        year.append(i.get('year') or 'NA')
        
        
        gen=i.get('genres')
        temp=[]
        for j in gen:
            temp.append(j.get('name') or 'NA')
        genres.append(temp)
            
        them_=i.get('themes')
        temp_1=[]
        for k in them_:
            temp_1.append(k.get('name') or 'NA')
        themes.append(temp_1)
        
        demographics.append(i.get('demographics')[0].get('name') if i.get('demographics') else 'NA')
        studio.append(i.get('studios')[0].get('name') if i.get('studios') else 'NA')
```

```python
dict_anime= {
    'rank': rank,
    'title': title,
    'title_english': title_english,
    
    'source': source,
    'type': type_2,
    'released_as': type_1,
    'genres': genres,
    'themes': themes,
    
    'episodes': episodes,
    'duration': duration,
    'rating': rating,
    'airing': airing,  
    'season': season,
    'status': status,
    
    'score': score,
    'votes': votes,
    'favourites': favourites,
    
    'year': year,
    'studio': studio,
    'demographics': demographics
    }

df = pd.DataFrame(dict_anime)
```
